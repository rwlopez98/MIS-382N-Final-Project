#Data Preperation

Since our data is majoritively categorical we're going to forgo any transformations/interpretation, and go ahead and just encode our categorical/ordinal columns with one-hot encoding.

```python
from imblearn.over_sampling import SMOTE

#Starting with conversion of non-continuous/non-binary categorical data to dummy data.
catColumns = ['GenHlth', 'MentHlth', 'PhysHlth', 'Age', 'Education', 'Income']
df = pd.get_dummies(df, columns=catColumns)

targetColumn = ['Diabetes_012']

X = df.drop(targetColumn, axis=1)
y = df[targetColumn]

#Applying SMOTE for balancing our minority classes
smote = SMOTE(random_state=42)
X, y = smote.fit_resample(X, y)
```

#Training and Validation Using K-Folds w/ Cross Validation

To improve the efficency of our training, testing, and validation. We are going to go ahead and leverage the K-Folds and Cross Validation Scoring tools from the sklearn library. While computationally expensive, we can leverage the computing power available to us to improve our work efficency and select a couple of the best performing classifier to dive into deeper.

```python
#Importing KFold
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
import numpy as np

k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=25)
scaler = StandardScaler()
models = []
trainedmodel = []

#Bringing in our models that we want to compare to each other.
from sklearn import tree
dt = tree.DecisionTreeClassifier()
dtPipeline = make_pipeline(scaler, dt)
models.append(dtPipeline)
trainedmodel.append(dt)

from sklearn import linear_model
lr = linear_model.LogisticRegression(max_iter = 1000)
lrPipeline = make_pipeline(scaler, lr)
models.append(lrPipeline)
trainedmodel.append(lr)

from sklearn import neural_network
mlp = neural_network.MLPClassifier(random_state = 25,max_iter = 1000)
mlpPipeline = make_pipeline(scaler, mlp)
models.append(mlpPipeline)
trainedmodel.append(mlp)

from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnbPipeline = make_pipeline(scaler, gnb)
models.append(gnbPipeline)
trainedmodel.append(gnb)

from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42, n_jobs=-1, verbose=0)
rfPipeline = make_pipeline(scaler, rf)
models.append(rfPipeline)
trainedmodel.append(rf)

from sklearn.decomposition import PCA
rfpca = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42, n_jobs=-1, verbose=0)
pca = PCA(n_components=0.95, random_state=42)
rfpcaPipeline = make_pipeline(scaler, pca, rfpca)
models.append(rfpcaPipeline)
trainedmodel.append(rfpca)

results = []

#Running our scoring for each model
for model in models:
  pos = 0
  scores = cross_val_score(model, X, y, cv=kf, scoring='accuracy')
  results.append([model, trainedmodel[pos], scores])
  print(model, np.mean(scores))
  pos += 1
```

## Takeaways
Based on our above results, we can see that our RF model without PCA, outperformed all of our other models that we tested on average across our K-Folds Cross Validation. Catching our own mistake, PCA is actually not an appropriate tool to be applying to catagorical data, and will not be considered moving forward.

## Visualization Tools

The visualization tools used in the remainder of our project to produce ouir graphics.

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix, roc_auc_score, roc_curve, auc
from sklearn.preprocessing import label_binarize
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

model_results = []  # List to store results

def evaluate_and_store(model, X_test, y_test, model_name="Model", cv_f1_score = None):
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = cv_f1_score

    # ROC AUC
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)
        y_test_bin = label_binarize(y_test, classes=[0.0, 1.0, 2.0])
        roc_auc = roc_auc_score(y_test_bin, y_proba, average='macro', multi_class='ovr')
    else:
        roc_auc = None

    model_results.append({
        "Model": model_name,
        "Accuracy": acc,
        "F1 Macro": f1,
        "ROC AUC": roc_auc,
        "Proba": y_proba if 'y_proba' in locals() else None
    })

    # You can still call your original evaluate_model if needed
    evaluate_model(model, X_test, y_test, title=model_name + " - Confusion Matrix")

def evaluate_model(model, X_test, y_test, title="Confusion Matrix"):
    y_pred = model.predict(X_test)

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print(f"Accuracy Score: {acc:.4f}")

    # Classification report
    print("Classification Report:\n", classification_report(y_test, y_pred))

    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion Matrix:\n", cm)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No Diabetes', 'Pre-Diabetes', 'Diabetes'],
                yticklabels=['No Diabetes', 'Pre-Diabetes', 'Diabetes'])
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title(title)
    plt.show()

    # ROC AUC (macro-average)
    if hasattr(model, "predict_proba"):
        y_proba = model.predict_proba(X_test)
        y_test_bin = label_binarize(y_test, classes=[0.0, 1.0, 2.0])
        auc_score = roc_auc_score(y_test_bin, y_proba, average='macro', multi_class='ovr')
        print(f"ROC AUC (Macro-average): {auc_score:.4f}")
    else:
        print("ROC AUC not available for this model (missing predict_proba).")


from sklearn.model_selection import train_test_split, cross_val_score
def run_cross_validation(model, X, y, cv=3):
    scores = cross_val_score(model, X, y, cv=cv, scoring='f1_macro')
    print(f"Cross-validated F1 Macro Scores: {scores}")
    print(f"Average F1 Macro Score: {np.mean(scores)}")
    return np.mean(scores)
```
